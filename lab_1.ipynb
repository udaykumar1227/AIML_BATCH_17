```python
import numpy as np
def sin(x):
    return np.sin(x)
def log(x):
    return np.log(x)
def log2(x):
    return np.log2(x)
def exp(x):
    return np.exp(x)
x = np.pi / 4
print("sin(pi/4) =", sin(x))
print("log(e) =", log(np.e))
print("log2(8) =", log2(8))
print("exp(1) =",exp(1))
```

```python
import matplotlib.pyplot as plt
people=['Varshini', 'Akhil', 'Lavanya', 'Mallesh']
age=[19, 17, 35, 43]
height=[135, 162, 176, 124]
weight=[45, 55, 65, 75]
plt.figure(figsize=(8, 6))
plt.scatter(age, height, c='pink', label='Height')
plt.scatter(age, weight, c='yellow', label='Weight')
plt.title('Scatter Plot of Age vs. Height and Weight')
plt.xlabel('Age')
plt.ylabel('Height and Weight')
plt.legend()
plt.grid(True)
plt.show()
# 4. Bar Chart
plt.figure(figsize=(8, 6))
bar_width=0.35
index=range(len(people))
plt.bar(index, height, width=bar_width, label='Height', color='pink')
plt.bar([i+bar_width for i in index], weight, width=bar_width, label='Weight', color='yellow')
plt.xlabel('People')
plt.ylabel('Height and Weight')
plt.title('Bar Chart of Height and Weight for Each Person')
plt.xticks([i+bar_width/2 for i in index], people)
plt.legend()
plt.grid(True)
plt.show()
# 5. Histogram
plt.figure(figsize=(8, 6))
plt.hist(height, bins=10, alpha=0.7, color='pink', label='Height')
plt.hist(weight, bins=10, alpha=0.7, color='yellow', label='Weight')
plt.title('Histogram of Height and Weight')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
